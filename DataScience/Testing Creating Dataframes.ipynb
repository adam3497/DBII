### 8.2 Operaciones en filas

```python
# calcular promedio de columna
df_exams['average'] = (df_exams['math score'] + df_exams['reading score'] + df_exams['writing score'] + df_exams['language score'])/4
df_exams.round(2)
df_exams
```

### 8.3 Método value_counts()

```python
# Contar los elementos en la columna "gender"

# función len
len(df_exams['gender'])

# método .count()
df_exams['gender'].count()
```

```python
# contar elementos "gender" por categoría
df_exams['gender'].value_counts()
```

```python
# obtener las frecuencias relativas (porcentaje)
df_exams['gender'].value_counts(normalize=True)
```

```python
# contar elementos "parental level of education" por categoria
df_exams['parental level of education'].value_counts()
```

```python
# frecuencia relativa de "parental level of education" redondeado a 2 decimales

# primera forma
np.round(df_exams['parental level of education'].value_counts(normalize=True), 2)

# segunda forma
df_exams['parental level of education'].value_counts(normalize=True).round(2)
```

## 9. Ordenar un DataFrame con sort_values()

```python
# ordenar por una columna
df_exams.sort_values('math score')
```

```python
# orden descendiente por una columna
df_exams.sort_values('math score', ascending=False)
```

```python
# orden descendiente por multiples columnas
df_exams.sort_values(['math score', 'reading score'], ascending=False)
```

```python
# orden descendiente por multiples columnas y actualizar dataframe
df_exams.sort_values(['math score', 'reading score'], ascending=False, inplace=True)
df_exams
```

```python
# ordenar descendiente con una función key
df_exams.sort_values('race/ethnicity', key=lambda col:col.str.lower())
```
